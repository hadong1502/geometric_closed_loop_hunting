```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
path = Path(r'C:\git\rot2-project\data\2024-01-23_latency_test\output0.txt')
lines = None
with open(path, 'r') as f:
    lines = f.readlines()
```

```python
store = {
    'wait': [],
    'd1': [],
    'd2': [],
    'd3': [],
    'd4': [],
    'snapshot': [],
    'imdecode': [],
    'analyze': [],
    'move': [],
}
for line in lines:
    if ': ' not in line:
        continue
    key, value = line.split(': ')
    if key in store.keys():
        store[key].append(value)
```

```python
fig, axs = plt.subplots(len(store), 1, figsize=(8, 4 * len(store)))

for i, key in enumerate(store.keys()):
    store[key] = np.array(store[key], dtype=np.float32)
    axs[i].hist(store[key])
    axs[i].set_title(key)

plt.tight_layout()
plt.show()
```
