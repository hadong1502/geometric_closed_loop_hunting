```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

## Video 2

```python
is_red_path = Path('/Users/yefan/Desktop/rot2/rot2-project/data/2024-01-24_latency_test1/is_red2.npy')
is_red = np.load(is_red_path)
y_loc_path = Path('/Users/yefan/Desktop/rot2/rot2-project/data/2024-01-24_latency_test1/y_loc2.npy')
y_loc = np.load(y_loc_path)

print(is_red.shape)
print(y_loc.shape)
```

```python
n_frames = is_red.shape[0]
plt.plot(((y_loc - np.mean(y_loc)) / np.std(y_loc)))
plt.plot(is_red)
```

```python
y_loc_diff = np.diff(y_loc)
y_loc_diff = y_loc
is_red_diff = np.diff(is_red)
is_red_plot = is_red
start, end = 1000, 2000

plt.plot(((y_loc_diff - np.mean(y_loc_diff)) / np.std(y_loc_diff))[start:end])
plt.plot((is_red_plot[start:end] - 0.5) * 2)
```

```python
rising_idx = np.where(is_red_diff == 1)[0]
print(rising_idx)
falling_idx = np.where(is_red_diff == -1)[0]
print(falling_idx)
```

```python
y_loc_norm = (y_loc_diff - np.mean(y_loc_diff)) / np.std(y_loc_diff)
high_idx = y_loc_norm > 1
low_idx = y_loc_norm < -1

a = 3000
b = 4000
plt.plot(high_idx[a:b])
plt.plot(low_idx[a:b])
```

```python
high_diff = np.diff(high_idx.astype(int))
low_diff = np.diff(low_idx.astype(int))
plt.plot(high_diff[a:b])
plt.plot(low_diff[a:b])
```

```python
high_falling = np.where(high_diff == -1)[0]
low_rising = np.where(low_diff == -1)[0]
print(len(high_falling))
print(len(low_rising))
print(len(rising_idx))
print(len(falling_idx))
```

```python
camera_fps = 193.57
s_per_frame = 1 / camera_fps
print(high_falling)
print(falling_idx[:len(high_falling)])
high_latency = high_falling - falling_idx[:len(high_falling)]
plt.hist(high_latency * s_per_frame, bins=100)
plt.ylabel('Count')
plt.xlabel('Latency (s)')
```

```python
print(low_rising)
print(rising_idx[:len(low_rising)])
high_latency = low_rising - rising_idx[:len(low_rising)]
plt.hist(high_latency * s_per_frame, bins=100)
plt.ylabel('Count')
plt.xlabel('Latency (s)')
```
