```python
mouse_pos = df['mouse_pos']
mouse_pos[0]
```

```python
def get_mouse_pos(mouse_pos):
    mouse_pos = mouse_pos[1:-1].replace(' ', '')
    x, y, _ = mouse_pos.split('.')
    mouse_pos = np.array([int(x), int(y)])
    return mouse_pos
```

```python
def visualize_mouse_pos(path):
    df = pd.read_csv(path)
    mouse_pos = df['mouse_pos']
    mouse_pos_int = mouse_pos.apply(get_mouse_pos)
    mouse_x_pos = np.array(mouse_pos_int.apply(lambda x: x[0]))
    mouse_y_pos = np.array(mouse_pos_int.apply(lambda x: x[1]))
    return mouse_x_pos, mouse_y_pos
```

```python
mouse_pos_int = mouse_pos.apply(get_mouse_pos)
mouse_pos_int[0]
```
